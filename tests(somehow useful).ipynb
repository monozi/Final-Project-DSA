**LINKED** **LIST**

```python
import tkinter as tk

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insert(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def display(self):
        current = self.head
        while current is not None:
            print(current.data)
            current = current.next

class LinkedListGUI(tk.Frame):
    def __init__(self, master):
        super().__init__(master)

        self.linked_list = LinkedList()

        # Create a text box to display the linked list
        self.linked_list_text_box = tk.Text(self, width=50, height=10)
        self.linked_list_text_box.pack()

        # Create a button to insert a new node into the linked list
        self.insert_button = tk.Button(self, text="Insert", command=self.insert_node)
        self.insert_button.pack()

    def insert_node(self):
        # Get the data to insert into the linked list
        data = self.linked_list_text_box.get("1.0", "end-1c")

        # Insert the data into the linked list
        self.linked_list.insert(data)

        # Update the text box to display the linked list
        self.display_linked_list()

    def display_linked_list(self):
        self.linked_list_text_box.delete("1.0", "end")
        self.linked_list.display()

if __name__ == "__main__":
    root = tk.Tk()
    linked_list_gui = LinkedListGUI(root)
    linked_list_gui.pack()

    root.mainloop()
```

**STACK**

```python
import tkinter as tk

class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.items:
            raise Exception("Stack is empty")
        return self.items.pop()

    def peek(self):
        if not self.items:
            raise Exception("Stack is empty")
        return self.items[-1]

class StackGUI(tk.Frame):
    def __init__(self, master):
        super().__init__(master)

        self.stack = Stack()

        # Create a text box to display the contents of the stack
        self.stack_text_box = tk.Text(self, width=50, height=10)
        self.stack_text_box.pack()

        # Create a button to push an element onto the stack
        self.push_button = tk.Button(self, text="Push", command=self.push_element)
        self.push_button.pack()

        # Create a button to pop an element from the stack
        self.pop_button = tk.Button(self, text="Pop", command=self.pop_element)
        self.pop_button.pack()

    def push_element(self):
        # Get the data to push onto the stack
        data = self.stack_text_box.get("1.0", "end-1c")

        # Push the data onto the stack
        self.stack.push(data)

        # Update the text box to display the stack
        self.display_stack()

    def pop_element(self):
        try:
            # Pop an element from the stack
            data = self.stack.pop()
        except Exception as e:
            # Display an error message if the stack is empty
            print(e)
            return

        # Update the text box to display the stack
        self.display_stack()

    def display_stack(self):
        self.stack_text_box.delete("1.0", "end")
        for item in self.stack.items[::-1]:
            self.stack_text_box.insert("end", f"{item}\n")

if __name__ == "__main__":
    root = tk.Tk()
    stack_gui = StackGUI(root)
    stack_gui.pack()

    root.mainloop()
```

**QUEUE**

```python
import tkinter as tk

class Queue:
    def __init__(self):
        self.items = []

    def enqueue(self, item):
        self.items.append(item)

    def dequeue(self):
        if not self.items:
            raise Exception("Queue is empty")
        return self.items.pop(0)

    def peek(self):
        if not self.items:
            raise Exception("Queue is empty")
        return self.items[0]

class QueueGUI(tk.Frame):
    def __init__(self, master):
        super().__init__(master)

        self.queue = Queue()

        # Create a text box to display the contents of the queue
        self.queue_text_box = tk.Text(self, width=50, height=10)
        self.queue_text_box.pack()

        # Create a button to enqueue an element onto the queue
        self.enqueue_button = tk.Button(self, text="Enqueue", command=self.enqueue_element)
        self.enqueue_button.pack()

        # Create a button to dequeue an element from the queue
        self.dequeue_button = tk.Button(self, text="Dequeue", command=self.dequeue_element)
        self.dequeue_button.pack()

        # Create a button to peek at the element at the front of the queue
        self.peek_button = tk.Button(self, text="Peek", command=self.peek_element)
        self.peek_button.pack()

    def enqueue_element(self):
        # Get the element to enqueue from the user
        element = self.queue_text_box.get("1.0", "end-1c")

        # Enqueue the element onto the queue
        self.queue.enqueue(element)

        # Update the text box to display the queue
        self.display_queue()

    def dequeue_element(self):
        try:
            # Dequeue an element from the queue
            element = self.queue.dequeue()
        except Exception as e:
            # Display an error message if the queue is empty
            print(e)
            return

        # Update the text box to display the queue
        self.display_queue()

    def peek_element(self):
        try:
            # Peek at the element at the front of the queue
            element = self.queue.peek()
        except Exception as e:
            # Display an error message if the queue is empty
            print(e)
            return

        # Display the element in the text box
        self.queue_text_box.delete("1.0", "end")
        self.queue_text_box.insert("end", f"{element}\n")

    def display_queue(self):
        self.queue_text_box.delete("1.0", "end")
        for element in self.queue.items:
            self.queue_text_box.insert("end", f"{element}\n")

if __name__ == "__main__":
    root = tk.Tk()
    queue_gui = QueueGUI(root)
    queue_gui.pack()

    root.mainloop()
```

**TREE**

```python
import tkinter as tk

class Tree:
    def __init__(self):
        self.root = None

    def create_node(self, data):
        new_node = Node(data)
        return new_node

    def insert(self, node, parent, child_index):
        if parent is None:
            self.root = node
        else:
            if child_index < len(parent.children):
                parent.children.insert(child_index, node)
            else:
                parent.children.append(node)

    def delete(self, node):
        if node == self.root:
            self.root = None
        else:
            parent = node.parent
            index = parent.children.index(node)
            parent.children.remove(node)

    def search(self, data):
        current = self.root
        while current is not None:
            if current.data == data:
                return current
            elif data < current.data:
                current = current.children[0]
            else:
                current = current.children[-1]
        return None

class TreeGUI(tk.Frame):
    def __init__(self, master):
        super().__init__(master)

        self.tree = Tree()

        # Create a text box to display the contents of the tree
        self.tree_text_box = tk.Text(self, width=50, height=10)
        self.tree_text_box.pack()

        # Create a button to insert a node into the tree
        self.insert_button = tk.Button(self, text="Insert", command=self.insert_node)
        self.insert_button.pack()

        # Create a button to delete a node from the tree
        self.delete_button = tk.Button(self, text="Delete", command=self.delete_node)
        self.delete_button.pack()

        # Create a button to search for a node in the tree
        self.search_button = tk.Button(self, text="Search", command=self.search_node)
        self.search_button.pack()

    def insert_node(self):
        # Get the node data from the user
        node_data = self.tree_text_box.get("1.0", "end-1c")

        # Insert the node into the tree
        self.tree.insert(Node(node_data))

        # Update the text box to display the tree
        self.display_tree()

    def delete_node(self):
        # Get the node data to delete from the user
        node_data_to_delete = self.tree_text_box.get("1.0", "end-1c")

        # Delete the node from the tree
        self.tree.delete(node_data_to_delete)

        # Update the text box to display the tree
        self.display_tree()

    def search_node(self):
        # Get the node data to search for from the user
        node_data_to_search_for = self.tree_text_box.get("1.0", "end-1c")

        # Search for the node in the tree
        node = self.tree.find(node_data_to_search_for)

        # Display the node data in the text box if it is found
        if node is not None:
            self.tree_text_box.delete("1.0", "end")
            self.tree_text_box.insert("end", f"{node.data}\n")
        else:
            print(f"Node {node_data_to_search_for} not found.")

    def display_tree(self):
        self.tree_text_box.delete("1.0", "end")
        current = self.tree.root
        while current is not None:
            self.tree_text_box.insert("end", f"{current.data}\n")
            current = current.children[0]

if __name__ == "__main__":
    root = tk.Tk()
    tree_gui = TreeGUI(root)
    tree_gui.pack()

    root.mainloop()
```

**HASH TABLE**

```python
import tkinter as tk

class HashTable:
    def __init__(self):
        self.table = {}

    def insert(self, key, value):
        self.table[key] = value

    def delete(self, key):
        del self.table[key]

    def search(self, key):
        return self.table.get(key, None)

class HashTableGUI(tk.Frame):
    def __init__(self, master):
        super().__init__(master)

        self.hash_table = HashTable()

        # Create a text box to display the contents of the hash table
        self.hash_table_text_box = tk.Text(self, width=50, height=10)
        self.hash_table_text_box.pack()

        # Create a button to insert a value into the hash table
        self.insert_button = tk.Button(self, text="Insert", command=self.insert_value)
        self.insert_button.pack()

        # Create a button to delete a value from the hash table
        self.delete_button = tk.Button(self, text="Delete", command=self.delete_value)
        self.delete_button.pack()

        # Create a button to search for a value in the hash table
        self.search_button = tk.Button(self, text="Search", command=self.search_value)
        self.search_button.pack()

    def insert_value(self):
        # Get the key and value from the user
        key = self.hash_table_text_box.get("1.0", "end-1c")
        value = self.hash_table_text_box.get("1.0", "end-1c")

        # Insert the value into the hash table
        self.hash_table.put(key, value)

        # Update the text box to display the hash table
        self.display_hash_table()

    def delete_value(self):
        # Get the key of the value to delete from the user
        key_to_delete = self.hash_table_text_box.get("1.0", "end-1c")

        # Delete the value from the hash table
        self.hash_table.delete(key_to_delete)

        # Update the text box to display the hash table
        self.display_hash_table()

    def search_value(self):
        # Get the key of the value to search for from the user
        key_to_search_for = self.hash_table_text_box.get("1.0", "end-1c")

        # Search for the value in the hash table
        value = self.hash_table.get(key_to_search_for)

        # Display the value in the text box if it is found
        if value is not None:
            self.hash_table_text_box.delete("1.0", "end")
            self.hash_table_text_box.insert("end", f"{value}\n")
        else:
            print(f"Value for key {key_to_search_for} not found.")

    def display_hash_table(self):
        self.hash_table_text_box.delete("1.0", "end")
        for key, value in self.hash_table.items():
            self.hash_table_text_box.insert("end", f"{key}: {value}\n")

if __name__ == "__main__":
    root = tk.Tk()
    hash_table_gui = HashTableGUI(root)
    hash_table_gui.pack()

    root.mainloop()
```
